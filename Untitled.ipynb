```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data1 = pd.read_csv('original.csv')
```

```python
data1.head()
```

```python
data1
```

```python
data1 = data1.drop('name', axis = 1)
```

```python
data1 = data1.replace({'origin':{1:'America',2:'Asia',3:'Europe'}})
```

```python
data1.head()
```

```python
data1
```

```python
df = pd.get_dummies(data1,columns=['origin'])
```

```python
df.head()
```

```python
X = df.drop(['mpg','origin_Europe'],axis=1)
```

```python
y = df['mpg']
```

```python
X_train,X_test,y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =1)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
linearRegression = LinearRegression().fit(X_train,y_train)
```

```python
linearRegression.score(X_train,y_train)
```
